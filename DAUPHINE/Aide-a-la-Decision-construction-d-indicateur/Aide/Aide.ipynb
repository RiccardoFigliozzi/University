**Calcul du poids**

```
def poids(index):
  diff = len(new_data) - new_data[index].isnull().sum()
  return diff/len(new_data)
```

```
poids('African Development Bank CPIA')
```

```
list_poi = []
for index in nom[6:]:
  #print("Index:", index)
  #print("Poids:", poids(index))
  list_poi.append((index, float(poids(index))))
```

```
list_poi
```

```
new_data = new_data.fillna(0)
new_data
```

```
def avg_pon(pays):
  num = 0
  den = 0
  note = new_data[new_data['Country'] == pays]
  for i in range(len(list_poi)):
    num += float(note.iloc[0][i+1]* list_poi[i][1])
    den += list_poi[i][1]
    
  return num/den
```

```
avg_pon('Denmark')
```

```
new_data['new_score'] = 0
for i in range(len(new_data)):
  new_data['new_score'][i] = avg_pon(new_data.iloc[i]["Country"])  
```
