```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
class_name = 'Occupancy'
df1 = pd.read_csv('datatraining.txt', skipinitialspace=True, na_values='?', 
                 keep_default_na=True)
df2 =pd.read_csv('datatest.txt', skipinitialspace=True, na_values='?', 
                 keep_default_na=True)
df3=pd.read_csv('datatest2.txt', skipinitialspace=True, na_values='?', 
                 keep_default_na=True)
df4=pd.merge(df1, df2, how='outer')
df=pd.merge(df4, df3, how='outer')
df.shape

```

```python
df['Day']=pd.to_datetime(df['date']).dt.day
df['Hour']=pd.to_datetime(df['date']).dt.hour
df['Month']=pd.to_datetime(df['date']).dt.month
df['Year']=pd.to_datetime(df['date']).dt.year
df['Data']=pd.to_datetime(df['date']).dt.normalize()
del df['date']
df.head()
```

```python
df['Day']=df['Data'].dt.dayofweek
df.drop(['Year','Month','Data'], axis='columns', inplace=True)
del df['HumidityRatio']
```

```python
from data_preparation import prepare_dataset
```

```python
res = prepare_dataset(df, class_name)
df, feature_names, class_values, numeric_columns, rdf, real_feature_names, features_map = res
df.head()
```

PARTITIONING

```python
from sklearn.model_selection import train_test_split, cross_val_score ,cross_val_predict

from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
attributes = [col for col in df.columns if col != class_name]
X = df[attributes].values
y = df[class_name]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

## knn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10, weights='distance')
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
predicted_probas_knn=knn.predict_proba(X_test)
```

## naive bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive = GaussianNB()
naive.fit(X_train, y_train)
```

```python
predicted_probas_naive=naive.predict_proba(X_test)
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis = LogisticRegression(random_state=0)
logis.fit(X_train, y_train)
```

```python
predicted_probas_logis=logis.predict_proba(X_test)
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree=DecisionTreeClassifier(ccp_alpha=0.0, class_weight=None, criterion='gini',
                       max_depth=9, max_features=None, max_leaf_nodes=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, presort='deprecated',
                       random_state=0, splitter='best')
tree.fit(X_train,y_train)
```

```python
predicted_probas_tree=tree.predict_proba(X_test)
```

## matrice con lift chart

```python
import matplotlib.pyplot as plt
import scikitplot as skplt
```

```python
# Set up a grid of plots
fig = plt.figure(figsize=(10, 10)) 
fig_dims = (2,2)

# Plot death and survival counts
#plt.subplot2grid(fig_dims, (0, 0))
skplt.metrics.plot_cumulative_gain(y_test,predicted_probas_knn,title="Cumulative Gains Curve with k-NN")

plt.show()

# Plot Pclass counts
#plt.subplot2grid(fig_dims, (0, 1))
skplt.metrics.plot_cumulative_gain(y_test,predicted_probas_naive,title="Cumulative Gains Curve with Naive Bayes")
plt.show()
# Plot Sex counts
#plt.subplot2grid(fig_dims, (1, 0))
skplt.metrics.plot_cumulative_gain(y_test,predicted_probas_logis,title="Cumulative Gains Curve with Logistic Regression")
plt.show()
# Plot Embarked counts
#plt.subplot2grid(fig_dims, (1, 1))
skplt.metrics.plot_cumulative_gain(y_test,predicted_probas_tree,title="Cumulative Gains Curve with Decision tree")


plt.show()
```
